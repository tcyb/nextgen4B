```python
# xmin, xmax = (1, len(mdf))

xmin, xmax = (90, 160)

letterorder = ['C', 'A', 'T', 'G']
colors = [[0.9,0.9,0.1],
          [0.9,0.1,0.9],
          [0.1,0.9,0.1],
          [0.1,0.1,0.9]]
ymax = 1
```

```python
x = mdf.index[xmin:xmax]
plt.figure(figsize=[12,10])
for i in range(len(letterorder)):
    plt.plot(x, mdf.iloc[x,i], color=colors[i])
    plt.fill_between(x, lbdf.iloc[x,i], ubdf.iloc[x,i], color=colors[i], alpha=0.5)
    plt.ylim(0,ymax)
plt.legend(letterorder)
    
plt.show()
```
